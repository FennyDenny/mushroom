```python
x=df1.drop("class",axis=1)
y=df1["class"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=12)
```

creating the train test split

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

By using DecisionTreeClassifier.  Got the accuracy as 1

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf')#It is a type of kernel we have aligned. Rbf is radial basis function.
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

By using SVC .Got accuracy as 0.99

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(classification_report(y_test,predsvc))
```

By using KNeighborsClassifier.  Got accuaracy as 0.999

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
rfc.score(x_train,y_train)
predrfc=rfc.predict(x_test)
print(accuracy_score(y_test,predrfc))
print(classification_report(y_test,predrfc))
```

By using RandomForest Classifier .Got accuracy as 1

Got the accuracy in RandomForestClassifier and DecisionTreeClassifier as 1 . Got accuracy of SVC and KNN classifiers as 0.99 & 0.999 respectivley.  To check the best accuracy for which model, going to find cross validation score

# Hyper parameter tunning

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
parameter={'max_depth':np.arange(2,10),
          'criterion':['gini','entropy'],
          'splitter':['best','random'],
          'max_features':['auto','sqrt','log2']}
```

created the parameter list

```python
GCV=GridSearchCV(DecisionTreeClassifier(),parameter,cv=5)
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

These are the best parameter forms.  

```python
final_model=DecisionTreeClassifier(criterion='gini',max_depth= 9, max_features='sqrt',splitter= 'best')
final_model.fit(x_train,y_train)
pred=final_model.predict(x_test)
accuracy=accuracy_score(y_test,pred)
print (accuracy*100)
```

So here got the best model with DecisionTreeClassifier with accuracy 100.  We used the parmeters criterion='gini',max_depth= 9, max_features='sqrt',splitter= 'best'
